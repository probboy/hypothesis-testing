```python
def populate_country_temp(country):
    url = f'http://climatedataapi.worldbank.org/climateweb/rest/v1/country/cru/tas/year/{country}'
    response = requests.get(url)
    data = response.json()
    
    x = [x['year'] for x in data]
    y = [x['data'] for x in data]

    def update_db(value, table_name, year):
        INSERT_STR = f'UPDATE {table_name} ' 
        VALUE_STR = f"""SET temp_{country} = {value}"""
        WHERE_STR = f'WHERE year = {year}'
        INSERT_QUERY = INSERT_STR + VALUE_STR + WHERE_STR
    
        c.execute(INSERT_QUERY)
    
        conn.commit()  
    
    for (p,q) in zip(x,y):
        update_db(q,'temp',p)
```

```python
c.execute("""select * from fema1""")
df_fema = pd.DataFrame(c.fetchall())
df_fema.columns = [x[0] for x in c.description]
df_fema
```

```python
df_fema['period'] = df_fema.decade.apply(lambda x: x>(1950+2010)/2)
```

```python
df_fema['incident'] = 1
```

```python
df_fema.groupby('decade').sum().reset_index().loc[:,['decade','incident']]
```

```python
df_fema2 = df_fema.groupby('Year').sum().loc[:,['incident']].reset_index()
```

```python
df_fema2['decade'] = df_fema2.Year.apply(lambda x: x//10*10)
```

```python
df_fema3 = df_fema2.groupby('decade').mean().reset_index().loc[:,['decade','incident']]
```

```python
df_fema3
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
formula = 'incident ~ decade'
lm = ols(formula, df_fema3).fit()
table = sm.stats.anova_lm(lm, typ=2)
print(table)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
print(pairwise_tukeyhsd(df_fema3.incident,df_fema3.decade))
```
